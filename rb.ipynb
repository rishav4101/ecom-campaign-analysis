```python
import pandas as pd
import numpy as np
!pip install xlrd

data = pd.read_excel("Fashion - Assignment Data (3).xlsx")
data.columns = data.iloc[0]
data.drop([data.index[0]], inplace=True)
data = data.sort_values(by = 'timestamp')
data.reset_index(drop=True, inplace=True)
data.reset_index(inplace=True)
data.rename(columns = {'index':"order_id"}, inplace=True)
data
```

```python
from datetime import datetime

starttime = "2017-10-03 21:12:48"
endtime = "2017-10-01 23:06:18"	

df = (data['timestamp'] < datetime.strptime(starttime, "%Y-%m-%d %H:%M:%S")) & (data['timestamp'] > datetime.strptime(endtime, "%Y-%m-%d %H:%M:%S"))
fd = data.loc[df]
fd["hours"] = fd["timestamp"].dt.hour
fd["weekday"] = fd["timestamp"].dt.day_name()
fd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fd.groupby
```

```python
fd["total_prod"] = 0
for index, row in fd.iterrows():
    try:
        fd.loc[index,"Revenue"] = float(row["Revenue"])
    except:
        fd.loc[index,"Revenue"] = int(000)
    if isinstance(row['Number_of_Products'], int):
        fd.loc[index,"total_prod"] = row['Number_of_Products']
    else:
        s = 0
        for i in row['Number_of_Products'].split(','):
            try:
                s = s + int(i.encode("ascii"))
            except:
                s = s + int(i)
        fd.loc[index,"total_prod"] = s


fd
```

```python
# Total Revenue
print(sum(pd.to_numeric(fd["Revenue"], errors="ignore")))

# Total prod
fd.sum(axis=0)
```

```python
# Avg Revenue
ls = pd.to_numeric(fd["Revenue"], errors="coerce")
print(sum(ls)/len(ls))

# Avg Product
fd.mean(axis=0)
```
